```python
import pandas as pd
import matplotlib.pyplot as plt

training = 'archive/Uniqlo(FastRetailing) 2012-2016 Training - stocks2012-2016.csv'
test = 'archive/Uniqlo(FastRetailing) 2017 Test - stocks2017.csv'
df_train = pd.read_csv(training)
df_test = pd.read_csv(test)

df_train['Date'] = pd.to_datetime(df_train['Date'])

df_train = df_train.sort_values(by='Date')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_train['Date'], df_train['Close'], marker='o', color='b', label=' Price')
plt.title('Uniqlo Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (Yen)')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.bar(df_train['Date'], df_train['Volume'], color='skyblue')
plt.title('Uniqlo Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Trading Volume')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
